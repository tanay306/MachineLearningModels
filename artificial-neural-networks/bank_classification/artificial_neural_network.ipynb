```python
#Import Libraries

import numpy as np
import pandas as pd
import tensorflow as tf
from tabulate import tabulate
```

```python
#Read Dataset

dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
max_columns = len(X[0])
```

```python
print(y)
```

```python
#Gender column needs to be encoded

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
```

```python
print(X)
```

```python
#Geography column needs to be encoded

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
#Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Feature Scaling is needed to train the model

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Experimentation

table = [['Nodes in Layer 1', 'Nodes in Layer 2', 'Accuracy']]

from sklearn.metrics import accuracy_score
for i in range (4, max_columns+2):
    for j in range (4, i+1):
        ann = tf.keras.models.Sequential()
        ann.add(tf.keras.layers.Dense(units = i, activation = 'relu'))
        ann.add(tf.keras.layers.Dense(units = j, activation = 'relu'))
        ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
        ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
        ann.fit(X_train, y_train, batch_size = 32, epochs = 20)
        y_pred = ann.predict(X_test)
        y_pred = (y_pred > 0.5)
        acc = accuracy_score(y_test, y_pred)
        table.append([i,j,acc])
        
```

```python
print(tabulate(table))
```

```python
#Building the ANN
#After several experimentation the 
#Number of Hidden Layers required for the dataset are 2
#Number of Nodes in Layer 1 are 7
#Number of Nodes in Layer 2 are 5

ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units = 7, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 5, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
#Using our ANN model to predict if the customer with the following informations will leave the bank:

#Geography: France
#Credit Score: 600
#Gender: Male
#Age: 40 years old
#Tenure: 3 years
#Balance: $ 60000
#Number of Products: 2
#Does this customer have a credit card ? Yes
#Is this customer an Active Member: Yes
#Estimated Salary: $ 50000

#So, should we say goodbye to that customer ?

print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
```

```python
#Predicting the Test set results

y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Making the Confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
acc = accuracy_score(y_test, y_pred)
print("The Accuracy of the model is: ", acc)
```
